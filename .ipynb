```python
import pandas as pd
import pymysql
```

```python
raw_csv_data = pd.read_csv('lumulli_mtg_inventory.csv')
```

```python
df = raw_csv_data.copy()
df = df.drop(['Unnamed: 6', 'Unnamed: 7'], axis=1)
column_names = ['Owned','Name','Set','Rarity','Value','Total']
df.columns = column_names
df.Name = df.Name.str.replace("[!@#$%^&*,/'-]", '')
df.Value = df.Value.str.replace("[!@#$%^&*,/'-]", '')
df.Total = df.Total.str.replace("[!@#$%^&*,/'-]", '')
df = df.drop_duplicates(subset=['Name','Set'], keep='first')
df = df.dropna()
df = df.reset_index(drop=True)
df.Owned = df.Owned.astype(int)
df.Value = df.Value.astype(float)
df.Total = df.Total.astype(float)
df
```
